```python
# Multiple Linear Regression

# Importing the libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Importing the dataset
dataset = pd.read_csv('/content/drive/MyDrive/Refactor_Notes/ML_Models_Python_Sessions/Data_files/50_Startups.csv')
```

```python
dataset.head(5)
```

```python
## Selecting our target variable(dependent variable) and keeping the features aside (independent variable)
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, 4]
```

```python
X.head(5)
```

```python
y.head(5)
```

```python
## Refer the slides
```

```python
#Convert the column into categorical columns
## https://www.learndatasci.com/glossary/dummy-variable-trap/

## Refer the link for dummy variable trap
states=pd.get_dummies(X['State'], drop_first=True)

states.head(5)
```

```python
X.head(5)
```

```python

# Drop the state coulmn
X=X.drop('State',axis=1)

X.head(3)

```

```python
# concat the dummy variables
X=pd.concat([X,states],axis=1)
```

```python
X.head(5)
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
X_train.head(2)
```

```python
X_test.head(2)
```

```python
y_train.head(2)
```

```python
y_test.head(2)
```

```python
# Fitting Multiple Linear Regression to the Training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test set results
y_pred = regressor.predict(X_test)

```

```python
y_pred
```

```python
y_test
```

```python
## Calculate correlation between predicted and actual values
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
```
